```python
import geopandas as gpd
```

```python

path_to_data = gpd.datasets.get_path("nybb") #확장자 파일 geojson
gdf = gpd.read_file(path_to_data)

gdf
```

```python
gdf = gdf.set_index("BoroName")
```

```python
gdf["area"] = gdf.area
gdf["area"]
```

```python
gdf
```

```python
gdf["centroid"] = gdf.centroid
```

```python
gdf.plot("area", legend=True)
```

```python
ax = gdf["geometry"].plot()
gdf["centroid"].plot(ax=ax, color="yellow")
```
